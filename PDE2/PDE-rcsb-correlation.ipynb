```python
file="cry-binding.txt"
df =pd.read_csv(file, header=None, sep='\t', engine='python', )
df.columns = ["Compound", "Inhibition"]
```

```python
def rmse(pred, tar):
    return np.sqrt(((pred - tar) ** 2).mean())

def convertAffinity(x):
    return 8.314*300/1000/4.184*np.log(2/1000000)
```

```python
df1 = df.copy()
df1['Exp dG'] = df1['Inhibition'].apply(convertAffinity)

df1['Calc dG']=''
dG1_list=[]
for k in range(7):
    tmp = df1['Compound'][k]
    cmpd = str(tmp)
    file="/mnt/u1/glide-pde-conf1/poses/rcsb/amber/"+cmpd+"/pbsa/Results.dat"
    readFile = open(file, "r")
    for line in readFile:
        line = line.strip() # strip end-on-line
        if re.match(r"DELTA TOTAL", line):
            dG1_total = line.split()
            dG1_list.append(dG1_total[2])
df1['Calc dG'] = dG1_list
df1['Calc dG'] = pd.to_numeric(df1['Calc dG'])
```
