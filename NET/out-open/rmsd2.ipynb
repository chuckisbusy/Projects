```python
flist=['/mnt/u1/NET/new-net/model-2481/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-1493/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-1076/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-0685/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-0204/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-1649/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-1870/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-2138/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-2440/ptraj/rmsfit.backbone.out',
'/mnt/u1/NET/new-net/model-2455/ptraj/rmsfit.backbone.out']
```

```python
rmsd=[]
for file in flist:
    data=np.loadtxt(file,comments='#',usecols=1)
    rmsd.append(data)
```

```python
x = np.arange(len(rmsd[0]))/500
plt.plot(x,rmsd[0],label='Model 1',linewidth =1.5)
plt.plot(x,rmsd[1],label='Model 2',linewidth =1.5)
plt.plot(x,rmsd[2],label='Model 3',linewidth =1.5)
plt.plot(x,rmsd[3],label='Model 4',linewidth =1.5)
plt.plot(x,rmsd[4],label='Model 5',linewidth =1.5)
plt.plot(x,rmsd[5],label='Model 6',linewidth =1.5)
plt.plot(x,rmsd[6],label='Model 7',linewidth =1.5)
plt.plot(x,rmsd[7],label='Model 8',linewidth =1.5)
plt.plot(x,rmsd[8],label='Model 9',linewidth =1.5)
plt.plot(x,rmsd[9],label='Model 10',linewidth =1.5)
leg = plt.legend(fontsize=18,handlelength=1,loc="best")

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(4.0)

plt.ylabel("RMSD ($\AA$)", fontsize=50)
plt.yticks(fontsize=33)
plt.xlabel("Time (ns)", fontsize=50)
plt.xticks(fontsize=35)
plt.tick_params(axis='both',labelsize='38', length=9, width=5)
plt.tight_layout()
plt.savefig("../../figs/rmsd.png", transparent=True)
```

```python
def calcStats (data):
    avg=np.average(data)
    std=np.std(data, dtype=np.float64)
    result = avg, u"\u00B1", std
    print('The average RMSD for the backbone is:', result)
    #return result
    minRMSD = data[data > .01].min(axis=0)
    maxRMSD = data.max()
    print("Range of RMSD is ", minRMSD, "-", maxRMSD)
```

```python
calcStats(rmsd[0])
```

```python
x = np.arange(len(rmsd[1]))/100000

plt.plot(x,rmsd[0],label='RMSD',linewidth=2,color='black')

leg = plt.legend(fontsize=35,handlelength=1,loc="best")

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(8.0)

plt.ylabel("RMSD ($\AA$)", fontsize=50)
plt.yticks((np.arange(0, 25, step=5)), fontsize=35)
plt.xlabel("Time ($\mu$s)", fontsize=50)
plt.xticks(fontsize=35)
plt.tick_params(axis='both', labelsize='37', length=9, width=5)
plt.tight_layout()
plt.savefig("../../figs/rmsd1.png", transparent=True)
```
