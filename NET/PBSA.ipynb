```python
def calcStats (data):
    avg=np.average(data)
    std=np.std(data, dtype=np.float64)
    result = avg, u"\u00B1", std
    print('The average is:', result)
    #return result
    minRMSD = data.min(axis=0)
    maxRMSD = data.max()
    print("Range is ", minRMSD, "-", maxRMSD)
```

```python
dataocc=[]
dataopen=[]

dataocc = np.loadtxt("MMPBSA_dG_occ.mmpbsa")
dataopen = np.loadtxt("MMPBSA_dG_open.mmpbsa")

```

```python
sns.kdeplot(np.array(dataocc),  label='NET Occ', color='blue', linewidth='5')
sns.kdeplot(np.array(dataopen), label='NET Open', color='red', linewidth='5')

leg = plt.legend(fontsize=25,handlelength=1,loc="best")

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(8.0)
    
plt.tick_params(which='both',direction='in',length=8, width=3,)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.xlabel('Binding Energy (kcal/mol)', fontsize=25)
plt.ylabel('Probability', fontsize=25)
plt.title('MMPBSA',fontsize=35)
plt.savefig("../figs/MMPBSA.png", transparent=True, bbox_inches='tight')
```
