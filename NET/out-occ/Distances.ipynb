```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib import rcParams
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 10,10
rcParams['axes.linewidth'] = 4.0
rcParams['savefig.dpi'] = 500
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
rcParams['agg.path.chunksize'] = 10000
```

```python
data=[]
flist=["dist.e159-r597ne.out", "dist.e159-r597n2.out", "dist.y495-k594.out", "dist.d493o1-k563.out", "dist.d493o2-k563.out", 
       "dist.d325n1-r600o1.out", "dist.d325n2-r600o2.out", "dist.k151-d549.out", "dist.e162-r597n1.out", 
       "dist.e162-r597n2.out", "dist.r163-e546o1.out", "dist.r163-e546o2.out", "dist.e324-r601.out", 
       "dist.k35-p562.out", "dist.y31-k563.out", "dist.t491O-k585N.out"]
for i in flist:
    rootdir="/mnt/u1/NET/tat-net-out-occ/net-tat-lipid"
    datafile= rootdir + "/ptraj/" + i
#    print(['{}'.format(i)])
    datatmp=np.loadtxt(datafile,comments='#',usecols=1)
    data.append(datatmp)
```

```python
def calcStats (data):
    avg=np.average(data)
    std=np.std(data, dtype=np.float64)
    result = avg, u"\u00B1", std
    print('The average Dist is:', result)
    #return result
    minVal = data[data > .01].min(axis=0)
    maxVal = data.max()
    print("Range of Dist is ", minVal, "-", maxVal)

def HBondCutoff (data, cutoffTemp):    
    allLen = len(data)
    cutoff=(cutoffTemp + .1)
    t = list(filter((cutoff).__ge__, data))
    lenVar = len(t)
    percent = (lenVar/allLen * 100)
    print("Percentage for a", cutoffTemp ,"\u212B cuttoff is: ", percent,"%")
```

```python
x = np.arange(len(data[1]))/100000

plt.plot(x,data[0], label="O1(D212)-Ne(R650)" ,color="r")
plt.plot(x,data[1], label="O2(D212)-N2(R650)" ,color="k")

leg = plt.legend(fontsize=25,handlelength=1,loc="best")

# set the linewidth of each legend object
for legobj in leg.legendHandles:
    legobj.set_linewidth(8.0)

plt.ylabel("RMSD ($\AA$)", fontsize=35)
plt.yticks((np.arange(0, 26, step=5)), fontsize=35)
plt.xlabel("Time ($\mu$s)", fontsize=35)
plt.xticks(fontsize=35)
plt.tick_params(axis='both', labelsize='35', length=7, width=5)
plt.tight_layout()
#plt.savefig("../../figs/dist1.png", transparent=True)

print("Stats for D212 - R650")
calcStats(data[0])
calcStats(data[1])
```
