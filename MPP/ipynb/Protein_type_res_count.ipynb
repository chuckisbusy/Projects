```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

## Import data

```python
plt.rcParams['axes.linewidth'] = 2.0
plt.rcParams['savefig.dpi'] = 500
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['axes.axisbelow'] = True
```

```python
df=pd.read_csv("/home/cadeniran/ipn/data/64protein-type.csv", header=0, sep=',', engine='python')
selectd = pd.read_csv("/home/cadeniran/ipn/data/selectd.csv", header=0, sep=',', engine='python')
```

# Analyze protein type

```python
print(df['type_id'].value_counts())
df_per = df.drop(df[df['type_id'] == 'Peripheral'].index)
df_tm = df.drop(df[df['type_id'] == 'Transmembrane'].index)
```

## Plot Residue Count

```python
plt.rcParams['figure.figsize'] = 17,7

# Plot the RMSDs
X=df_per['id']
Y1=df_per['res_count']

# Plot
plt.scatter(X, Y1, marker="o", s=10**2, c='black')
plt.axhline(y = 400, color = 'r', linestyle = 'dashed')
plt.grid(True)
plt.xlabel("Protein", fontsize=25)
plt.ylabel("Residue count", fontsize=25)
plt.tick_params(axis='x', rotation=90, labelsize=15)
plt.tick_params(axis='y', labelsize=20)

# Save figure
plt.savefig("/home/cadeniran/ipn/figs/perscatter.png", format='png',bbox_inches='tight')
```

```python
plt.rcParams['figure.figsize'] = 17,7

# Plot the RMSDs
X=df_tm['id']
Y1=df_tm['res_count']

# Plot
plt.scatter(X, Y1, marker="o", s=10**2, c='black')
plt.axhline(y = 400, color = 'r', linestyle = 'dashed')
plt.grid(True)
plt.rc('axes', axisbelow=True)
plt.xlabel("Protein", fontsize=25)
plt.ylabel("Residue count", fontsize=25)
plt.tick_params(axis='x', rotation=90, labelsize=15)
plt.tick_params(axis='y', labelsize=20)

# Save figure
plt.savefig("/home/cadeniran/ipn/figs/tmscatter.png", format='png',bbox_inches='tight')
```
