```python
import os
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 17,7
plt.rcParams['axes.linewidth'] = 2.0
plt.rcParams['savefig.dpi'] = 500
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

### Import Data

```python
os.chdir("/home/cadeniran/u1/amphiscan-pep/")
df = pd.read_csv("MCC-MCCAT.csv", header=None, sep=',', engine='python', names=['Peptide', 'MCC', 'MCC_AT'])
df['MCC'] = pd.to_numeric(df['MCC'], errors='coerce')
```

### Make Plot

```python
X=df["Peptide"]
Y1=df["MCC"]
Y2=df["MCC_AT"]

# Plot
fig,ax=plt.subplots()
ax.plot(X, Y1, marker="o", markersize=10, c='red')
ax.plot(X, Y2, marker="o", markersize=10, c='black')  
ax.grid(True)
ax.set_xlabel("Peptide", fontsize=25)
ax.set_ylabel("MCC", fontsize=25)
ax.tick_params(axis='x', rotation=90, labelsize=15)
ax.tick_params(axis='y', labelsize=20)
plt.ylim(-0.6, 1)

# Save figure
plt.savefig("/home/cadeniran/ipn/figs/MCC.png", format='png',bbox_inches='tight')
```
