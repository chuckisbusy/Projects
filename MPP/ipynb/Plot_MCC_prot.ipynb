```python
import os
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 17,7
plt.rcParams['axes.linewidth'] = 2.0
plt.rcParams['savefig.dpi'] = 500
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

### Import Data

```python
os.chdir("/home/cadeniran/u1/amphiscan-64prot/")
df = pd.read_csv("MCC-MCCAT.csv", header=0, sep=',', engine='python', names=['Peptide', 'MCC', 'MCC_AT'])
df['MCC'] = pd.to_numeric(df['MCC'], errors='coerce')
```

### Make Plot

```python
X=df["Peptide"]
Y1=df["MCC"]
Y2=df["MCC_AT"]

# Plot
plt.plot(X, Y1, marker="o", markersize=10, c='red', label="MCC")
plt.plot(X, Y2, marker="o", markersize=10, c='black', label="MCC-AT")

leg = plt.legend(fontsize=20,handlelength=1,loc="best")

# get the individual lines inside legend and set line width
for line in leg.get_lines():
    line.set_linewidth(4)

plt.grid(True)
plt.xlabel("Protein PDB ID", fontsize=25)
plt.ylabel("MCC", fontsize=25)
plt.tick_params(axis='x', rotation=90, labelsize=15)
plt.tick_params(axis='y', labelsize=20)
plt.ylim(-0.5, 1)

# Save figure
plt.savefig("/home/cadeniran/ipn/figs/MCC.png", format='png',bbox_inches='tight')
```
