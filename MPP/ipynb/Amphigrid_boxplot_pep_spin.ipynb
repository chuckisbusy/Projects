## Plot single boxplot

```python
plt.rcParams['figure.figsize'] = 7,7
ax = sns.boxplot(x="pert_num", y="RMSD", data=df_long["0.1_325"], showfliers = False, linewidth=2)
ax = sns.swarmplot(x="pert_num", y="RMSD", data=df_long["0.1_325"], color="r", alpha=0.5)

ax.set_ylim([0, 26])
plt.xlabel("Perturbation #", fontsize=25)
plt.ylabel("RMSD", fontsize=25)
plt.tick_params(axis='both', labelsize='17', length=9, width=3)
```
