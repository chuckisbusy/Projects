```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
```

```python
# define the smiles string separately and covert it into a molecule sturcture
volinanserin = Chem.MolFromSmiles('COC1=CC=CC(=C1OC)[C@@H](C2CCN(CC2)CCC3=CC=C(C=C3)F)O') #volinanserin
ketanserin = Chem.MolFromSmiles('C1CN(CCC1C(=O)C2=CC=C(C=C2)F)CCN3C(=O)C4=CC=CC=C4NC3=O') #ketanserin
serotonin = Chem.MolFromSmiles('C1=CC2=C(C=C1O)C(=CN2)CCN') #sert
```

```python
# combine all molecules into 1 object
mols = [Chem.MolFromSmiles(smi) for smi in ('COC1=CC=CC(=C1OC)[C@@H](C2CCN(CC2)CCC3=CC=C(C=C3)F)O',
                                          'C1CN(CCC1C(=O)C2=CC=C(C=C2)F)CCN3C(=O)C4=CC=CC=C4NC3=O',
                                          'C1=CC2=C(C=C1O)C(=CN2)CCN')]
# Define mols names
molNames=["volinanserin","ketanserin","serotonin"]
```

```python
for i in range(3):
    mols[i].SetProp('Name', molNames[i])
```
