```python
import os
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 7,7
plt.rcParams['axes.linewidth'] = 2.0
plt.rcParams['savefig.dpi'] = 500
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
np.set_printoptions(legacy='1.25')
```

# Plot RMSD 

```python
file="/home/cadeniran/storage/bwhiteside/5HT/for_Ben/MD/newDM5/5-production/rmsd.agr"
data=np.loadtxt(file,comments='@',usecols=1)
```

```python
x = np.arange(len(data))/2
plt.plot(x,data,label='Backbone',color='red')

plt.ylabel("RMSD ($\AA$)", fontsize=30)
plt.yticks(fontsize=28)
plt.xlabel("Time (ns)", fontsize=30)
plt.xticks(np.arange(0, 550, 100), fontsize=28)
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
plt.tight_layout()
plt.savefig("/home/cadeniran/ipn/figs/rmsd.png", transparent=True)
```

# Plot RMSD & Score DM5

```python
file="/home/cadeniran/storage/bwhiteside/5HT/for_Ben/Docking/run_docking_trunc_5HT2A/output_files/all_extracted_pdbs/out_top_models/score_vs_RMSD.csv"
dfdm5=pd.read_csv(file, header=None, sep=' ', engine='python', names=['RMSD', 'Score'])
```

```python
X=dfdm5["RMSD"]
Y=dfdm5["Score"]

# Plot
plt.title("Score vs RMSD: DM5", fontdict={'fontsize': 30, 'fontweight': 'bold', 'color': 'black'})
plt.scatter(X, Y, marker='s')
plt.xlabel("RMSD ($\AA$)", fontsize=40)
plt.ylabel("Score (REU)", fontsize=40)
plt.tick_params(axis='x', labelsize=28, length=7, width=3)
plt.xticks(np.arange(0, 8.5, 1))
#plt.xlim(-0.5, 7.5)
plt.tick_params(axis='y', labelsize=28, length=7, width=3)
plt.ylim(-20, 30)
#plt.yticks(np.arange(-15, 30, 10))
plt.axhline(y = 0, color = 'black', linestyle = 'dashed', linewidth=4)

# Save figure
plt.savefig("/home/cadeniran/ipn/figs/score-dm5.png", format='png',bbox_inches='tight')
```
