```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fileRMSD = "/net/share/cad234/projects/p2x/atp/ptraj/rms.protein-backbone-prod1.rmsfit"
data = np.loadtxt(fileRMSD)
X2 = (data[:,0])/250
plt.ylabel("RMSD $\AA$")
plt.xlabel("Time (ns)")
plt.plot(X2,data[:,1], label="Run 1")
plt.legend(loc='upper left')
plt.gcf().savefig("rmsd-atp.png",dpi=300)
```

```python
fileRMSF = "/net/share/cad234/projects/p2x/atp/ptraj/rmsf_prod1.out"
data = np.loadtxt(fileRMSF)
plt.ylabel("RMSF $\AA$")
plt.xlabel("Residues")
plt.plot(data[:,0],data[:,1], label="Run 1")
plt.legend(loc='upper left')
plt.gcf().savefig("rmsf-atp.png",dpi=300)
```
