```python
import matplotlib.pylab as plt
import numpy as np
from matplotlib import rcParams
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 8,8
rcParams['axes.linewidth'] = 4.0
rcParams['savefig.dpi'] = 300
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

```python
file="/net/share/cad234/projects/p2x/analysis/pca/project.dat"
data=np.loadtxt(file,usecols=(0,1,2,3),comments="#")
```

```python
xray_x=data[:,1][0:2]
xray_y=data[:,2][0:2]
apo_x=data[:,1][2:3927]
apo_y=data[:,2][2:3927]
```

```python
plt.scatter(xray_x,xray_y,s=8.0,color="k",alpha=1.0,label='X-ray struc',linewidth=15)
plt.scatter(apo_x,apo_y,s=3.0,color="steelblue",alpha=1.0,label='P2X Apo')
plt.legend(fontsize=18,markerscale=3,scatterpoints=1)
plt.xlabel('PC1 ($\AA$)', fontsize=25)
plt.ylabel('PC2 ($\AA$)', fontsize=25)
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
#plt.title("PC1 vs PC2",fontsize=35)
plt.tight_layout()
plt.savefig("pca-p2x-apo.pdf", transparent=True)
```
