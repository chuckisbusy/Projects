```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 10,10
rcParams['axes.linewidth'] = 4.0
rcParams['savefig.dpi'] = 500
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

```python
for i in [1,2,3]:
        fileRMSD = "/net/share/cad234/projects/p2x/apo/rms/rms.protein-backbone-prod{0}.rmsfit".format(i)
        data = np.loadtxt(fileRMSD)
        X2 = (data[:,0])/2.5
        plt.plot(X2,data[:,1], label="Run {0}".format(i))
plt.legend(loc='upper left', fontsize=20)
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
plt.ylabel("RMSD ($\AA$)", fontsize=25)
plt.xlabel("Time (ns)", fontsize=25)
plt.savefig("rmsd.pdf", transparent=True)
```

```python
for i in [1, 2, 3]:
        fileRMSF = "/net/share/cad234/projects/p2x/apo/rms/rmsf_prod{0}.out".format(i)
        data = np.loadtxt(fileRMSF)
        plt.plot(data[:,0],data[:,1], label="Run {0}".format(i), linewidth='2')
        
plt.legend(loc='upper left', fontsize=20)
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
plt.ylabel("RMSF ($\AA$)", fontsize=25)
plt.xlabel("Residue Numbers", fontsize=25)
plt.savefig("rmsf.pdf", transparent=True)
```
