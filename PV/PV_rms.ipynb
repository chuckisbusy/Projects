```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 10,10
rcParams['axes.linewidth'] = 4.0
rcParams['savefig.dpi'] = 500
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

```python
cases=["apv_1RWY_Ca","apv_1RWY_Mg","apv_1S3P_Ca","apv_1S3P_Mg","apv_1XVJ_Ca","apv_1XVJ_Mg","bpv_1G33_Ca","bpv_1G33_Mg","bpv_1RRO_Ca","bpv_1RRO_Mg"]
loc="/net/share/cad234/projects/pv/"
```

```python
rmsd=[]
for i in np.arange(10):
    datafile= loc + cases[i] + "/rmsf/rmsd_rmsf-prod1.dat"
    data=np.loadtxt(datafile,comments='#',usecols=1)
    rmsd.append(data)
```

```python
x = np.arange(len(rmsd[0]))/5
plt.plot(x,rmsd[0],label=r'$\alpha$PV Ca$^{2+}$',color='lightsalmon')
plt.plot(x,rmsd[1],label=r'$\alpha$PV Mg$^{2+}$',color='cyan')
plt.plot(x,rmsd[2],label=r'$\alpha$PV S55D/E59D Ca$^{2+}$',color='red')
plt.plot(x,rmsd[3],label=r'$\alpha$PV S55D/E59D Mg$^{2+}$',color='steelblue')
plt.plot(x,rmsd[4],label=r'$\alpha$PV D94S/G98E Ca$^{2+}$',color='tomato')
plt.plot(x,rmsd[5],label=r'$\alpha$PV D94S/G98E Mg$^{2+}$',color='lightblue')

plt.plot(x,rmsd[6],label=r'$\Delta \beta$PV Ca$^{2+}$',color='firebrick')
plt.plot(x,rmsd[7],label=r'$\Delta \beta$PV Mg$^{2+}$',color='mediumblue')
plt.plot(x,rmsd[8],label=r'$\beta$PV Ca$^{2+}$',color='darkred')
plt.plot(x,rmsd[9],label=r'$\beta$PV Mg$^{2+}$',color='darkblue')

plt.ylabel("RMSD $\AA$", fontsize=30)
plt.yticks(fontsize=28)
plt.xlabel("Time (ns)", fontsize=30)
plt.xticks(fontsize=28)
plt.legend(fontsize=11,handlelength=2,loc="best")
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
plt.tight_layout()
#plt.title(r'RMSD of $\alpha$- and $\beta$-PV', fontsize=30)
#plt.savefig("rmsd.jpeg", transparent=True)
```

```python
rmsf=[]
for i in np.arange(10):
    datafile= loc + cases[i] + "/rmsf/rmsf_prod1.out"
    data=np.loadtxt(datafile,comments='#',usecols=1)
    rmsf.append(data)
```

```python
plt.plot(np.arange(len(rmsf[0])),rmsf[0],label=r'$\alpha$PV Ca$^{2+}$',color='red',linestyle='dashed',markersize=6)
plt.plot(np.arange(len(rmsf[1])),rmsf[1],label=r'$\alpha$PV Mg$^{2+}$',color='blue',linestyle='dashed',markersize=6)
plt.plot(np.arange(len(rmsf[2])),rmsf[2],label=r'$\alpha$PV S55D/E59D Ca$^{2+}$',color='red',linestyle='solid',linewidth='4',markersize=10)
plt.plot(np.arange(len(rmsf[3])),rmsf[3],label=r'$\alpha$PV S55D/E59D Mg$^{2+}$',color='blue',linestyle='solid',linewidth='4',markersize=10)
plt.plot(np.arange(len(rmsf[4])),rmsf[4],label=r'$\alpha$PV D94S/G98E Ca$^{2+}$',color='red',linestyle='solid',linewidth='1',markersize=8)
plt.plot(np.arange(len(rmsf[5])),rmsf[5],label=r'$\alpha$PV D94S/G98E Mg$^{2+}$',color='blue',linestyle='solid',linewidth='1',markersize=8)

plt.plot(np.arange(len(rmsf[6]))+36,rmsf[6],label=r'$\Delta \beta$PV Ca$^{2+}$',color='darkred',linestyle='dashed',markersize=6)
plt.plot(np.arange(len(rmsf[7]))+36,rmsf[7],label=r'$\Delta \beta$PV Mg$^{2+}$',color='darkblue',linestyle='dashed',markersize=6)
plt.plot(np.arange(len(rmsf[8])),rmsf[8],label=r'$\beta$PV Ca$^{2+}$',color='darkred',linewidth='4',markersize=10)
plt.plot(np.arange(len(rmsf[9])),rmsf[9],label=r'$\beta$PV Mg$^{2+}$',color='darkblue',linewidth='4',markersize=10)

plt.ylabel("RMSF $\AA$", fontsize=30)
plt.yticks(fontsize=28)
plt.xlabel("Residues", fontsize=30)
plt.xticks(fontsize=28)
plt.legend(fontsize=11,handlelength=2,loc="best")
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
plt.tight_layout()
#plt.title(r'RMSF of $\alpha$- and $\beta$-PV', fontsize=30)
#plt.savefig("rmsf.pdf", transparent=True)
```

```python
plt.plot(np.arange(len(rmsf[0])),rmsf[0],label=r'$\alpha$PV Ca$^{2+}$',color='red',linestyle='dashed',markersize=6)
plt.plot(np.arange(len(rmsf[1])),rmsf[1],label=r'$\alpha$PV Mg$^{2+}$',color='blue',linestyle='dashed',markersize=6)

plt.ylabel("RMSF $\AA$", fontsize=30)
plt.yticks(fontsize=28)
plt.xlabel("Residues", fontsize=30)
plt.xticks(fontsize=28)
plt.legend(fontsize=16,handlelength=2,loc="best")
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
plt.title(r'RMSF of WT $\alpha$PV Ca and Mg', fontsize=30)
plt.tight_layout()
#plt.savefig("rmsf1.pdf", transparent=True)
```
