```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fileRMSD = "rms.protein-backbone-prod1.rmsfit"
data = np.loadtxt(fileRMSD)
X2 = (data[:,0])/500
plt.ylabel("RMSD $\AA$")
plt.xlabel("Time (ns)")
plt.plot(X2,data[:,1])
plt.title(r'RMSD $\alpha$-PV 1XVJ Ca')
plt.gcf().savefig("/net/share/cad234/projects/pv/figs/rmsd-1xvj.png",dpi=300)
```
