```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statistics
from pylab import rcParams
from scipy import stats
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 10,10
rcParams['axes.linewidth'] = 4.0
rcParams['savefig.dpi'] = 500
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

```python
cases=["apv_1RWY_Ca", "apv_1RWY_Mg", "apv_1S3P_Ca", "apv_1S3P_Mg", "apv_1XVJ_Ca", "apv_1XVJ_Mg","bpv_1G33_Ca", "bpv_1G33_Mg",  "bpv_1RRO_Ca", "bpv_1RRO_Mg"]
loc="/net/share/cad234/projects/pv/"
```

```python
casedata=[]
for i in np.arange(10):
    datafile= loc + cases[i] + "/watershell/ws-ca1.out"
    data=np.loadtxt(datafile,comments='#',usecols=1)
    casedata.append(data)
```

```python
sns.kdeplot(np.array(casedata[0]), bw=0.7, label=r'$\alpha$PV Ca$^{2+}$', color='red', linestyle='dashed',markersize=6)
sns.kdeplot(np.array(casedata[1]), bw=0.7, label=r'$\alpha$PV Mg$^{2+}$', color='blue', linestyle='dashed',markersize=6)
sns.kdeplot(np.array(casedata[2]), bw=0.7, label=r'$\alpha$PV S55D/E59D Ca$^{2+}$', color='red', linestyle='solid',linewidth='4',markersize=10)
sns.kdeplot(np.array(casedata[3]), bw=0.7, label=r'$\alpha$PV S55D/E59D Mg$^{2+}$', color='blue', linestyle='solid',linewidth='4',markersize=10)
sns.kdeplot(np.array(casedata[4]), bw=0.7, label=r'$\alpha$PV D94S/G98E Ca$^{2+}$', color='red', linestyle='solid',linewidth='1',markersize=8)
sns.kdeplot(np.array(casedata[5]), bw=0.7, label=r'$\alpha$PV D94S/G98E Mg$^{2+}$', color='blue', linestyle='solid',linewidth='1',markersize=8)

sns.kdeplot(np.array(casedata[6]), bw=0.7, label=r'$\Delta \beta$PV Ca$^{2+}$', color='darkred', linestyle='dashed',markersize=6)
sns.kdeplot(np.array(casedata[7]), bw=0.7, label=r'$\Delta \beta$PV Mg$^{2+}$', color='darkblue', linestyle='dashed',markersize=6)
sns.kdeplot(np.array(casedata[8]), bw=0.7, label=r'$\beta$PV Ca$^{2+}$', color='darkred', linestyle='solid',linewidth='4',markersize=10)
sns.kdeplot(np.array(casedata[9]), bw=0.7, label=r'$\beta$PV Mg$^{2+}$', color='darkblue', linestyle='solid',linewidth='4',markersize=10)
plt.legend(fontsize=14,handlelength=2,loc="best")
plt.tick_params(which='both',direction='in',length=8, width=3,)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
#plt.title("Water Count", fontsize=25)
plt.xlabel('# of waters', fontsize=25)
plt.ylabel('Probability', fontsize=25)
plt.tight_layout()
plt.savefig("watercount.pdf", transparent=True)
```

```python
for i in np.arange(10):
    sigma = statistics.mean(casedata[i])
    mu = stats.sem(casedata[i], axis=None, ddof=1)
    print("Average and Std Err for %s is: %0.2f and %0.4f" % (cases[i],sigma,mu))
```
