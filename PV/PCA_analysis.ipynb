```python
def plotfigure(data,color,area=0.1,alpha=0.5):
    x_axis=data[:,1]
    y_axis=data[:,2]

    plt.scatter(x_axis,y_axis,s=area,color=color,alpha=alpha)
    plt.ylim([-40,40])
    plt.xlim([20,100])
    plt.xlabel("PC1")
    plt.ylabel("PC2")
```

```python
# draw the integrated PC1  probablity of 
from scipy import spatial # which will be used find the nearst point

    
def calc_integral(center,axis,data,interval=10):
    low_bound = [center - interval]
    up_bound = [center + interval]
 #   print (low_bound, up_bound)
    data_array = np.asarray(axis)
#    print (data_array)

    
    low_index = min(range(len(axis)),key = lambda i: abs(axis[i] - low_bound)) #
    up_index = min(range(len(axis)), key = lambda i: abs(axis[i] - up_bound))#
    #print (axis[low_index],axis[up_index])
    
    a = 0
    for i in range(low_index,up_index):
        a += data[i]
    return a
            
```
