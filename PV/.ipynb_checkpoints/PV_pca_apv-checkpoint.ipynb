```python
import matplotlib.pylab as plt
import numpy as np
from matplotlib import rcParams
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 10,11
rcParams['axes.linewidth'] = 4.0
rcParams['savefig.dpi'] = 500
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

```python
file="/net/share/cad234/projects/pv/analysis/pca/apv/project.dat"
data=np.loadtxt(file,usecols=(0,1,2,3),comments="#")
```

```python
wtca_x=data[:,1][0:20000]
wtca_y=data[:,2][0:20000]
wtmg_x=data[:,1][19999:40000]
wtmg_y=data[:,2][19999:40000]
pv1ca_x=data[:,1][39999:60000]
pv1ca_y=data[:,2][39999:60000]
pv1mg_x=data[:,1][59999:80000]
pv1mg_y=data[:,2][59999:80000]
pv2ca_x=data[:,1][79999:100000]
pv2ca_y=data[:,2][79999:100000]
pv2mg_x=data[:,1][99999:120000]
pv2mg_y=data[:,2][99999:120000]
```

```python
plt.scatter(wtca_x,wtca_y,s=3.0,color="green",alpha=1.0,label=r'$\alpha$PV Ca$^{2+}$')
plt.scatter(wtmg_x,wtmg_y,s=3.0,color="red",alpha=1.0,label=r'$\alpha$PV Mg$^{2+}$')
plt.scatter(pv1ca_x,pv1ca_y,s=3.0,color="blue",alpha=1.0,label=r'$\alpha$PV S55D/E59D Ca$^{2+}$')
plt.scatter(pv1mg_x,pv1mg_y,s=3.0,color="purple",alpha=1.0,label=r'$\alpha$PV S55D/E59D Mg$^{2+}$')
plt.scatter(pv2ca_x,pv2ca_y,s=3.0,color="orangered",alpha=1.0,label=r'$\alpha$PV D94S/G98E Ca$^{2+}$')
plt.scatter(pv2mg_x,pv2mg_y,s=3.0,color="black",alpha=1.0,label=r'$\alpha$PV D94S/G98E Mg$^{2+}$')
plt.legend(fontsize=18,markerscale=6,scatterpoints=1)
plt.xlabel('PC1 ($\AA$)', fontsize=25)
plt.ylabel('PC2 ($\AA$)', fontsize=25)
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
plt.title("PC1 vs PC2 \u03B1PV WT & Mutants",fontsize=35)
plt.savefig("PCA.pdf", transparent=True, bbox_inches='tight')
```
