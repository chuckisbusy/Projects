```python
import numpy as np
import matplotlib.pylab as plt
from pylab import rcParams
from matplotlib.colors import LinearSegmentedColormap
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 20,14
rcParams['savefig.dpi'] = 300
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

```python
file1="/net/share/cad234/projects/pv/apv_1RWY_Ca/cmap/nonnative.cmap"
file2="/net/share/cad234/projects/pv/apv_1XVJ_Ca/cmap/nonnative.cmap"
data1=np.genfromtxt(file1,comments="#",usecols=[2])
data2=np.genfromtxt(file2,comments="#",usecols=[2])
data=np.subtract(data2,data1)
rdata = data.reshape(109,109)


df=rdata-np.min(rdata)
df=((df/np.max(df))*2)-1

position=np.arange(0,110,8)
#print(position)
xdistance=position+1
ygr=position+1

vmax = 1
cmap = LinearSegmentedColormap.from_list('mycmap',[(0, 'b'),(0.33, 'aqua'),(0.66, 'yellow'),(1, 'r')])

plt.imshow(df,cmap=cmap,interpolation="none",vmin=-1, vmax=1)
plt.xlabel("Residue ID", fontsize=35)
plt.ylabel("Residue ID", fontsize=35)
plt.xticks(position,xdistance,fontsize=30)
plt.title("$\u0394Contact = Contact_{\u03B1PV\ 1XVJ\ Ca} - Contact_{\u03B1PV\ 1RWY\ Ca}$",fontsize=35)
plt.yticks(position,ygr,fontsize=30)
ax = plt.gca()
ax.set_ylim(ax.get_ylim()[::-1])
cax=plt.colorbar().ax
cax.tick_params(labelsize=30)
plt.savefig("diff-cmap1.pdf", transparent=True)
```
