```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 10,10
rcParams['axes.linewidth'] = 4.0
rcParams['savefig.dpi'] = 500
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

```python
cases=["apv_1RWY_Ca","bpv_1RRO_Ca"]
loc="/net/share/cad234/projects/pv/"
```

```python
rmsf=[]
for i in np.arange(2):
    datafile= loc + cases[i] + "/rmsf/rmsf_prod1-nh.out"
    data=np.loadtxt(datafile,comments='#')
    rmsf.append(data)
```

```python
plt.plot(rmsf[0][:,0],rmsf[0][:,1],label=r'$\alpha$PV Ca$^{2+}$',color='red',linewidth='4',markersize=10)
plt.plot(rmsf[1][:,0],rmsf[1][:,1],label=r'$\beta$PV Ca$^{2+}$',color='darkred',linewidth='4',markersize=10)

plt.ylabel("RMSF $\AA$", fontsize=30)
plt.yticks(fontsize=28)
plt.xlabel("Residues", fontsize=30)
plt.xticks(fontsize=28)
plt.legend(fontsize=15,handlelength=2,loc="best")
plt.tick_params(axis='both',labelsize='25', length=7, width=5)
plt.tight_layout()
#plt.title(r'RMSF of $\alpha$- and $\beta$-PV', fontsize=30)
plt.savefig("diffrmsf1.pdf", transparent=True)
```
