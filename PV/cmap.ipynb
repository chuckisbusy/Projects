```python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image
from pylab import rcParams
from matplotlib.colors import LinearSegmentedColormap
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 20,14
rcParams['savefig.dpi'] = 500
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

```python
file1="/u1/cad234/pv/analysis/apv_1RWY_ca/cmap/nonnative.cmap"
data=np.genfromtxt(file1,comments="#",usecols=[2])
print(np.min(data))
print(np.max(data))
```

```python
data = data.reshape(109,109)
df=data-np.min(data)
df=((df/np.max(df))*2)-1
#print (data)
position=np.arange(0,110,8)
#print(position)
xdistance=position+1
ygr=position+1
print(np.max(df))
print(np.min(df))
vmax = 1
cmap = LinearSegmentedColormap.from_list('mycmap',[(0, 'w'),(0.33, 'aqua'),(0.66, 'yellow'),(1, 'r')])

plt.imshow(df,cmap=cmap,interpolation="none",vmin=-1, vmax=1)
plt.xlabel("Residue ID", fontsize=35)
plt.ylabel('Residue ID', fontsize=35)
plt.xticks(position,xdistance,fontsize=30)
plt.title('\u03B1PV 1RWY Ca WT',fontsize=35)
plt.yticks(position,ygr,fontsize=30)
ax = plt.gca()
ax.set_ylim(ax.get_ylim()[::-1])
cax=plt.colorbar().ax
cax.tick_params(labelsize=30)
plt.savefig("cmap1.pdf", transparent=True)
```
